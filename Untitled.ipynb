```python
import os
import cv2
import pandas as pd

# Directory containing your image dataset
dataset_dir = 'pc_parts/cables'

# Function to load images from directory
def load_images(directory):
    images = []
    for filename in os.listdir(directory):
        # Ensure we're only loading image files
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img_path = os.path.join(directory, filename)
#             # Read image using OpenCV
            img = cv2.imread(img_path)
#             cv2.imshow('Image', img)
#             # You may want to resize or preprocess the images here
            images.append(img)
#         print(img_path)
#         print(images)
    return images

# Load images
image_data = load_images(dataset_dir)

# Convert image data to DataFrame
data = {'image': image_data}
print(data)
df = pd.DataFrame(data)
df.head()

```
